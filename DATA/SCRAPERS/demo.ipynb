```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import sqlite3
import time
import pandas as pd
```

```python
import abc
from enum import Enum
```

```python
from dataclasses import dataclass
```

```python
class SQLITE_DATA_TYPES(Enum):
    NULL= "NULL"
    INTEGER= "INTEGER"
    REAL= "REAL"
    TEXT="TEXT"
    BLOB= "BLOB"
```

```python
@dataclass
class sqlTable:
    column: str
    dtype: SQLITE_DATA_TYPES
    primaryKey: bool
```

```python
class SQLITETable(abc.ABC):
    def __init__(self, con: sqlite3.Connection, tableName, columns: list(sqlTable)):
        self.columns = columns
        self.tableName = tableName
        self.con = con

    def createIfNotExists(self):
        cur = self.con.cursor()
        columnString = ""
        for index, t in enumerate(self.columns):
            columnString += "{t.column} {t.dtype.value}"
            if t.primaryKey:
                columnString += "PRIMARY KEY"
            if index != len(t.columns) - 1:
                columnString += ","

        cur.execute(
            "CREATE TABLE IF NOT EXISTS {self.tableName} ({columnString})")
        cur.close()
        return
        
```

```python
class NepseScraper():

    def __init__(self):
        self.driver = webdriver.Chrome(
            service=ChromeService(ChromeDriverManager().install()))
        self.baseURL = "https://www.nepalstock.com/"
        self.driver.get(self.baseURL)
        self.securities = []

    def scrapeTable(table, nextBtn):
        tableData = []
        while True:
            tableData.append(pd.read_html(table)[0])
            if 'disabled' in nextBtn.get_attribute('class'):
                break
            time.sleep(1)
            nextBtn.click()
        return tableData

    def getAllSecurities(self):
        page = self.driver.get(self.baseURL + "company")
        tableData = []
        nextBtn = self.driver.find_element(By.CLASS_NAME, 'pagination-next')
        while True:
            tableData.append(pd.read_html(self.driver.find_element(By.TAG_NAME, 'table').get_attribute('outerHTML'))[0])
            if 'disabled' in nextBtn.get_attribute('class'):
                break
            time.sleep(0.5)
            nextBtn.click()
        return tableData
        
```
